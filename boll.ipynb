```python
#boll
stock='SHSE.601606'
start='2024-08-25'
end='2026-02-13'
```

```python
history_data = history(symbol=stock, frequency='1d', start_time=start,  end_time=end, fields='close,eob', adjust=ADJUST_PREV, df= True)
history_data['eob'] = history_data['eob'].dt.date  #优化日期格式
```

```python
history_data['ma']=[0]*len(history_data)
history_data['std']=[0]*len(history_data)
history_data['upper']=[0]*len(history_data)
history_data['lower']=[0]*len(history_data)
for i in range(len(history_data)):
    history_data['ma'] = history_data['close'].rolling(window=20).mean()
    history_data['std'] = history_data['close'].rolling(window=20).std()
    history_data['upper'] = history_data['ma'] + 2 * history_data['std']
    history_data['lower'] = history_data['ma'] - 2 * history_data['std']
```

```python
fig=px.line(history_data, x='eob', y=['close','ma', 'upper','lower'],color_discrete_sequence=['black','white','yellow','purple'])  
fig.show()
```
